# 2. 축

```python
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['font.family'] = 'HYGungSo-Bold' #Windows일때
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글폰트 사용시 마이너스 글자 꺠지는 현상 없앤다
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family':'HYGungSo-Bold','size':50})
```

```python
plt.plot(x,y)
plt.xlabel('X축',color='red', loc = 'right') #left, conter, right
plt.ylabel('Y축',color='#00aa00', loc = 'top') #top, center bottom
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.show()
```
