# 9. 막대그래프(심화)

```python
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['font.family'] = 'HYGungSo-Bold' #Windows일때
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글폰트 사용시 마이너스 글자 꺠지는 현상 없앤다
```

```python
labels = ['강백호', '서태용', '정대만'] #이름
values = [190, 187, 184] #키

plt.bar(labels, values) #막대 그래프
```

```python
plt.barh(labels,values)
plt.xlim(175,195)
```

```python
bar = plt.bar(labels,values)
bar[0].set_hatch('/') #막대그래프 안에 단조로움 없애기위해 /, x, ..을 넣어본다
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

```python
bar = plt.bar(labels, values)
plt.ylim(175,195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height()+0.3, values[idx], ha = 'center', color='blue')
```
