```python
plt.plot(df['지원번호'], df['영어'])
plt.plot(df['지원번호'], df['수학'])
plt.grid() #격자가 생긴다
```

```python
plt.plot(df['지원번호'], df['영어'], label = '영어')
plt.plot(df['지원번호'], df['수학'], label = '수학')
plt.legend(loc = 'best', ncol=1) #label을 입력해줘야 legend를 할 수 있다
plt.grid(axis = 'x', color = 'purple', alpha = 0.6, linestyle = '--') #x축에 대해서만 격자가 생긴다
```
