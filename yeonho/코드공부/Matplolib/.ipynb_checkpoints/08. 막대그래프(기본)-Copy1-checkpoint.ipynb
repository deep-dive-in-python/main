# 8. 막대 그래프(기본)

```python
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['font.family'] = 'HYGungSo-Bold' #Windows일때
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글폰트 사용시 마이너스 글자 꺠지는 현상 없앤다
```

```python
labels = ['강백호', '서태용', '정대만'] #이름
values = [190, 187, 184] #키
# plt.plot  : 꺾은선 그래프
plt.bar(labels, values) #막대 그래프
```

```python
labels = ['강백호', '서태용', '정대만'] #이름
values = [190, 187, 184] #키
# plt.plot  : 꺾은선 그래프
plt.bar(labels, values, color = 'r') #막대 그래프
```

```python
labels = ['강백호', '서태용', '정대만'] #이름
values = [190, 187, 184] #키

colors = ['r', 'g', 'b'] #각각의 막대에 다른 색깔 부여
plt.bar(labels, values, color = colors, alpha = 0.5) #막대 그래프 & alpha = 0.5 투명도 부여
```

```python
labels = ['강백호', '서태용', '정대만'] #이름
values = [190, 187, 184] #키

plt.bar(labels, values)
plt.ylim(175, 195) #구분이 잘 안가는 y축의 lim을 정해준다
```

```python
plt.bar(labels, values, width = 0.3)
plt.ylim(175,200)
```

```python
plt.bar(labels, values, width = 0.3)
plt.xticks(rotation = 45) #x축의 이름데이터 각도로 45도로 설정
plt.yticks(rotation = 45)
plt.ylim(175,200)
```

```python
labels = ['강백호', '서태용', '정대만'] #이름
values = [190, 187, 184] #키

ticks = ['1번학생','2번학생','3번학생']
plt.bar(labels,values, width = 0.5)
plt.xticks(labels, ticks, rotation = 45)
plt.show()
```
