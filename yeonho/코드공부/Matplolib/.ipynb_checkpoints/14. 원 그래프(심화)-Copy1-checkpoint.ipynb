# 14. 원 그래프(심화)

```python
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['font.family'] = 'HYGungSo-Bold' #Windows일때
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글폰트 사용시 마이너스 글자 꺠지는 현상 없앤다

```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'JavaScript', 'C#', 'C++', 'ETC']
plt.pie(values, labels = labels, autopct = '%.1f%%', startangle = 90, counterclock= False) #%를 붙여주고 싶으면 '%.1f'에 %%를 더 붙인다
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'JavaScript', 'C#', 'C++', 'ETC']
# colors = ['b', 'g', 'r', 'c', 'm', 'y']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']
plt.pie(values, labels = labels, autopct = '%.1f%%', startangle = 90, counterclock= False, colors = colors)
#%를 붙여주고 싶으면 '%.1f'에 %%를 더 붙인다
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'JavaScript', 'C#', 'C++', 'ETC']
# colors = ['b', 'g', 'r', 'c', 'm', 'y']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']

explode = [0.05] * 6
plt.pie(values, labels = labels, autopct = '%.1f%%', startangle = 90, counterclock= False, colors = colors, explode = explode)
#%를 붙여주고 싶으면 '%.1f'에 %%를 더 붙인다
plt.show()
```

```python
wedgeprops = {'width' : 0.5} #도넛모양의 그래프
plt.pie(values, labels = labels, 
        autopct = '%.1f%%', startangle = 90, counterclock= False, colors = colors, explode = explode, wedgeprops = wedgeprops)
plt.show()
```

```python
wedgeprops = {'width' : 0.5, 'edgecolor':'w', 'linewidth' : 5} #도넛모양의 그래프, 테두리 부분에 color 반영
plt.pie(values, labels = labels, 
        autopct = '%.1f%%', startangle = 90, counterclock= False, colors = colors, wedgeprops = wedgeprops)
plt.show()
#글자들이 너무 겹쳐셔 안좋아보인다
```

```python
#데이터가 10%이상일때만 글자를 보여주겠다
def custom_autopct(pct) : 
    if pct >=10 :
        return ('%.1f%%' % pct)
    else :
        '-'
        
# def cutom_autopct(pct) : 
#     return ('%.1f%%' % pct) if pct >= 10  else ''  이렇게 2줄에 표현할 수도 있다
# return '{:.0f}%'.format(pct) if pct >=10 else '' 

plt.pie(values, labels = labels, 
        autopct =custom_autopct, startangle = 90, counterclock= False, colors = colors, wedgeprops = wedgeprops, pctdistance = 0.8)
#pctdistance는 pct간의 거리를 나타낸다.
plt.show()
```

```python
import pandas as pd
df = pd.read_excel('C:/Users/sarah/PythonDataWorkspace/Pandas/score.xlsx')
df
```

```python
grp2 = df.groupby(['학교'])
grp2.size()
```

```python
grp2.size()['능남고']
grp2.size()['북산고']
```

```python
values = [grp2.size()['북산고'], grp2.size()['능남고']] #[5,3]
labels = ['북산고', '능남고']
```

```python
colors = ['y','g'] #도넛색
explode = [0.05] * 2 #도넛간의 간격
wedgeprops = {'width' : 0.3} #도넛의 wedge 길이

plt.pie(values, labels=labels, autopct = '%.0f%%', colors = colors, explode = explode, wedgeprops = wedgeprops)
plt.legend(loc = 'best', ncol=2)
plt.title('<소속 학교>')
plt.show()
```
