# 3. 범례(legend)

```python
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['font.family'] = 'HYGungSo-Bold' #Windows일때
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글폰트 사용시 마이너스 글자 꺠지는 현상 없앤다
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y, label='sdfsdf')
plt.legend(loc = 'center') 
```

```python
plt.plot(x,y, label='sdfsdf')
plt.legend(loc='upper right') #matplot legend 검색하여 참고
```

```python
plt.plot(x,y, label='sdfsdf')
plt.legend(loc='center right')
```

```python
plt.plot(x,y, label = '범례')
plt.legend(loc='best')
plt.legend(loc=(0.2, 0.5)) #x축, y축 기준 0.5위치(0~1사이)
```
