# 4. 스타일

```python
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['font.family'] = 'HYGungSo-Bold' #Windows일때
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글폰트 사용시 마이너스 글자 꺠지는 현상 없앤다
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, linewidth=5) #라인두께
```

```python
plt.plot(x,y, marker='o')
```

```python
plt.plot(x,y,marker='o',linestyle = 'None')
```

```python
plt.plot(x,y,marker='v')
```

```python
plt.plot(x,y,marker='v', markersize = 10)
```

## marker의 정의

```python
plt.plot(x,y,marker='X', markersize = 10) #matplotlib marker 홈페이지에서 검색
```

```python
plt.plot(x,y,marker='o', markersize =20, markeredgecolor = 'red')
```

```python
plt.plot(x,y,marker='v', markersize = 20, markerfacecolor = 'red', markeredgecolor = 'yellow')
```

## 선 스타일

```python
plt.plot(x,y, linestyle='-')
```

```python
plt.plot(x,y, linestyle=':') #matplotlib linestyle 검색하면 나온다
```

## 색깔

```python
plt.plot(x,y, color = '#ff0000')
```

```python
plt.plot(x,y, color = 'b')
```

```python
plt.plot(x,y, color = 'g') #matplotlib color 검색
```

# 포맷

```python
plt.plot(x,y, 'bv--') #color, marker, linestyle을 한꺼번에
```

```python
plt.plot(x,y, 'go') #color, marker, linestyle='None'
```

```python
plt.plot(x,y, marker='o', color='g', linestyle=':') #color, marker, linestyle='None'
```

# 축약어

```python
plt.plot(x,y, marker='o', mfc='red', ms = 10, mec = 'g', ls=':')
#mfc : markerfacecolor
#ms : markersize
#mec : markeredgecolor
#ls 축약어
#matplotlib markerfacecolor mfc 을 검색하면 축약어가 나옴
```

# 투명도

```python
plt.plot(x,y, marker='o', mfc='red', ms = 10, mec = 'g', ls=':', alpha = 0.3) #투명도 alpha
```

```python
## 그래프 크기
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10, 5), dpi = 50) #dpi는 해상도 , dots per inch, 확대
plt.plot(x,y)
```

## 그래프 배경색

```python
plt.figure(facecolor = '#d67ade') #RGB 값을 선택할 수 있음
plt.plot(x,y)
```
