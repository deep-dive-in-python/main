## 데이터 제외하기 dropna

```python
import numpy as np #수학 관련 자료 할떄 numpy 많이 쓴다
df['학교'] = np.nan 
```

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

```python
df.dropna() #전체 데이터 중에서 nan을 포함하는 데이터를 삭제
```

```python
df.dropna(inplace=True) #진짜 데이터프레임 반영
```

```python
df = pd.read_excel('score.xlsx', index_col = '지원번호')
```

## axis 속성 : index or columns  
## how : any or all 
#### (any 는 NaN이 하나라도 포함되어 있으면 dropna하고,  all은 모두가 NaN이어야 dropna 한다)

```python
df.dropna(axis = 'index', how = 'any')  #nan가 하나라도 있는 row 삭제
# 4,5번 행이 사라진다 
```

```python
df.dropna(axis = 'columns', how = 'any')  #nan가 하나라도 있는 row 삭제, any는 default값이라 정의 안해줘도 된다

# sw특기 column이 사라진다
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns', how = 'all')
#데이터 전체가 NaN인 경우에만 column 삭제, '학교' column 사라지고 'sw특기'는 살아있음(전부가 NaN이 아니기에)
```
