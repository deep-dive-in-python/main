# 남자 데이터 정의

```python
import pandas as pd
df_m = pd.read_excel('201108_201108_연령별인구현황_월간.xlsx',
                     skiprows = 3, index_col = '행정기관',usecols = 'B, E:Y')
#skiprows=3 : 첫 3개 row 공백은 쓸필요 없기 때문에 무시
#index_col='행정기관' : index column을 지정
#usecols = 'B, E:Y' : B열과 E~Y까지의 필요한 columns만 사용
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int) #1,195,951의 , 부분은 없애기
df_m.iloc[0]
```

```python
df_m #전국 데이터 iloc[0]의 , 가 없어진 것을 확인할 수 잇다.
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus']
```

```python
plt.figure(figsize=(10,7)) #그래프 크기
plt.barh(df_m.columns, df_m.iloc[0] //1000) #단위 : 천명기준으로 해보자
```

```python
plt.figure(figsize=(10,7)) #그래프 크기
print(plt.barh(df_m.columns, -df_w.iloc[0] //1000)) #단위 : 천명기준으로 해보자, barh이니까 x축 y축이 bar와 다르겠네
print(plt.barh(df_w.columns, df_w.iloc[0] //1000))
#'-'를 붙여주니 대칭된 형태로 보인다

plt.title('2011 대한민국 인구 피라미드')
plt.savefig('2011_인구피라미드.png', dpi=100)
plt.show()
```
